**Las eliminamos porque estas columnas generar√≠an variables extra en el modelo de regresi√≥n lineal que claramente tendr√≠an multicolinealidad perfecta y adem√°s generan un costo computacional extra**


```python
def detectar_columnas_duplicadas(data):
    """
    Recorre las columnas y detecta cu√°les son duplicadas (aunque tengan nombres distintos).
    Imprime las columnas originales y sus copias.
    """
    cols = list(data.columns)
    cols_unicas = []
    duplicados = {}

    for col in cols:
        # buscar todas las columnas iguales 
        iguales = [col_unica for col_unica in cols_unicas if data[col].equals(data[col_unica])]
        
        if iguales: 
            for col_unica in iguales:
                duplicados.setdefault(col_unica, []).append(col)
        else:
            cols_unicas.append(col)
            duplicados[col] = []

    # printss
    print("ORIGINAL -> COPIA")
    for original, copias in duplicados.items():
        if copias:
            print(f"{original} -> {', '.join(copias)}")

detectar_columnas_duplicadas(TrainData)
```

```python
def eliminar_columnas_duplicadas(data):
    """
    Recorre  las columnas y elimina las que sean duplicadas (aunque tengan nombres distintos).
    Retorna un DataFrame limpio con columnas √∫nicas.
    """
    cols = list(data.columns)
    cols_unicas = []
    
    for i, col in enumerate(cols):
        duplicada = False
        for col_unica in cols_unicas:
            if data[col].equals(data[col_unica]):  # compara valores exactos
                #print(f"Columna '{col}' es duplicada de '{col_unica}' ‚Üí eliminada")
                duplicada = True
                break
        if not duplicada:
            cols_unicas.append(col)
    
    return data[cols_unicas]

# Uso:
print("Shape Antes:", TrainData.shape)
TrainData = eliminar_columnas_duplicadas(TrainData)
print("Shape final:", TrainData.shape)
```

### Clasificar Nacionalidad

- En el enunciado se menciona que un jugador extranjero puede ocupar un cupo de extranjero. Es decir, su valor proviene de el cupo y no por razones culturales. 
- por ejemplo, un brasilero no vale m√°s que un boliviano porque los brasileros sean mejores.  
- Para este punto hemos decidido utilizar una notaci√≥n binaria
(1 = Europeo,
0 = No europeo)


```python
def clasificar_nacionalidad(nacionalidad):
  europeos = [
    "ALB","AND","ARM","AUT","AZE","BEL","BIH","BGR","BLR","CYP","CZE","DEN","DNK","ENG","EST","FRO","FIN","FRA","GEO","GER","DEU","GIB","GRE","GRC","HUN","ISL",
    "IRL","ISR","ITA","KAZ","KOS","LVA","LIE","LTU","LUX","MLT","MDA","MNE","MKD","NED","NLD","NIR","NOR","POL","POR","ROU","RUS","SMR","SCO","SRB","SVK","SVN",
    "ESP","SWE","SUI","CHE","TUR","UKR","WAL"]

  if pd.isna(nacionalidad):
      return pd.NA
  paises = nacionalidad.split()
  pais=paises[-1]
  # Clasificaci√≥n si es europeo retorna 1 si no retorna 0
  if pais in europeos:
      return 1
  else:
      return 0
    
TrainData['Nacionalidad'] = TrainData['Nacionalidad'].apply(clasificar_nacionalidad)
TrainData = TrainData.rename(columns={"Nacionalidad": "¬øEuropa?"})
# df_copy["Nacionalidad"].describe()
print(TrainData.loc[110])
```

### Clasificaci√≥n de las posiciones
Para hacer un modelo manejable, hemos decidido agrupar las posiciones de los jugadores en 4 categor√≠as. Esto nos permitir√° hacer modelos especializados por posici√≥n (calcular mejor seg√∫n las posiciones).
Para que funcione adecuadamente es mejor correr el kernel de nuevo
- Portero
- Defensa
- Mediocampista
- Delantero

```python
def clasificar_posicion(posicion):
    # Arreglos con las posiciones agrupadas
    portero=["GK"]
    defensa=["CB","LB","RB","WB"]
    mediocampista=["DM", "CM","AM","LM", "RM"]
    delantero=["FW", "LW", "RW"]
    
    # Valores nulos
    if pd.isna(posicion) or str(posicion).strip() == "":
        return np.nan

    # Tomar la primera posici√≥n en caso de que existan m√°s posiciones. Ej: "LW,RW", se toma "LW"
    primera = str(posicion).split(",")[0].strip().upper()

    # Clasificaci√≥n dependiendo de la posici√≥n
    if primera in portero:
        return "Portero"
    elif primera in defensa:
        return "Defensa"
    elif primera in mediocampista:
        return "Mediocampista"
    elif primera in delantero:
        return "Delantero"
    else:
        # Si no coincide con ninguna posici√≥n
        return np.nan 

# Ejemplo de uso:
#TrainData.info()
TrainData['Posicion'] = TrainData['Posicion'].apply(clasificar_posicion)
TrainData["Posicion"].describe()
print(TrainData.loc[1062])
```

### Formato de edad: con decimales para ser m√°s precisos

```python

def convertir_edad_dataframe(data):
    def convertir_edad(edad:str)->float:
        if isinstance(edad,str):
            edad = edad.strip().split("-")
            edad = int(edad[0]) + int(edad[1])/365
        return edad
    
    data["Edad"] = data["Edad"].apply(convertir_edad)
    return data
    

# Test aplicado a tu DataFrame
print("Antes:", TrainData["Edad"].head(10))  # primeras 10 edades crudas

TrainData = convertir_edad_dataframe(TrainData)
print("\nDespu√©s:", TrainData[["Edad"]].head(10))
```

### Cambiar Formato Market value

```python
def convertir_market_value(market_value):
    # Se revisa si es nulo o no
    if pd.isna(market_value):
        return pd.NA

    market_value = str(market_value).strip()
    # Se revisa en tal caso de que no cuente con el simbolo de Euros
    if "‚Ç¨" not in market_value:
        return pd.NA
    
    # Se revisa que tenga la estructura con una expresi√≥n regular si no la tiene lo manda como nulo
    market_value= re.search(r'(?i)(?:‚Ç¨|EUR)\s*([0-9]+(?:[.,][0-9]+)?)\s*([mk])?\s*$', market_value)
    if not market_value:
        return pd.NA

    # Se separa el n√∫mero del sufijo
    monton = market_value.group(1)
    multiplicador = market_value.group(2)

    # Si el sufijo es m se multiplica por un mill√≥n y si es k se multiplica por 1000
    if multiplicador=="m":
        return float(monton)*1000000
    elif multiplicador=="k":
        return float(monton)*1000
    else:
        return pd.NA

TrainData["market_value"] = TrainData["market_value"].apply(convertir_market_value)
TrainData["market_value"].describe()
```

```python
from scipy.stats import zscore

def replace_outliers_zscore(df, cols, threshold=3):
    df_clean = df.copy()
    for col in cols:
        # calcular z-score solo sobre valores no nulos
        z = zscore(df_clean[col].dropna())
        mask = abs(z) > threshold
        
        # indices de outliers
        outlier_idx = df_clean[col].dropna().index[mask]
        
        # reemplazar con NaN
        df_clean.loc[outlier_idx, col] = pd.NA
    return df_clean

# aplicar a todas las columnas num√©ricas
TrainData.shape
num_cols = TrainData.select_dtypes(include="number").columns
TrainData = replace_outliers_zscore(TrainData, num_cols, threshold=3)
TrainData.shape
```

## FUNCI√ìN DE AGREGACI√ìN: 
- Este es el paso m√°s importante del notebook.

```python
def ultimo_o_moda(grupo):
    # si hay al menos una fecha v√°lida
    if grupo["Dia_partido"].notna().any():
        idx = grupo["Dia_partido"].idxmax()
        return grupo.loc[idx, "market_value"]
    else:
        # tomar moda del market_value
        moda = grupo["market_value"].mode()
        return moda.iloc[0] if not moda.empty else None

df_ultimo_valor = (
    TrainData.groupby("Jugador")
             .apply(ultimo_o_moda)
             .reset_index(name="market_value")
)

# Promedios y agregaciones
df_promedios = (
    TrainData.drop(columns=["market_value", "Dia_partido"])
      .groupby("Jugador")
      .agg({
        '¬øEuropa?': 'first',
        'Posicion': lambda x: x.mode().iloc[0],
        'Edad': 'max',
        'Goles': 'mean',
        'Tiros Totales': 'mean',
        'xG': 'mean',
        'npxG': 'mean',
        'xAG': 'mean',
        'Acciones_que_crean_tiros': 'mean',
        'Pases_intentados': 'mean',
        'Pases_progresivos': 'mean',
        'Regates_exitosos': 'mean',
        'Pases_medios_completados': 'mean',
        'Pases_largos_completados': 'mean',
        'xA': 'mean',
        'Pases_en_ultimo_tercio': 'mean',
        'Pases_balon_vivo': 'mean',
        'Pases_balon_muerto': 'mean',
        'Pases_al_hueco': 'mean',
        'Pases_centros': 'mean',
        'Pases_completados': 'mean',
        'Pases_fuera_de_juego': 'mean',
        'Valla_no_vencida': 'mean',
        'Errores_defnsivos_ocasion_tiro': 'mean',
        'Toques_en_zona_ofensiva': 'mean',
        'Toques_en_area_rival': 'mean',
        'Veces_que_fue_barrido_regate': 'mean',
        'Acarreos_ultimo_tercio': 'mean',
        'Malos_controles': 'mean',
        'Perdida_balon': 'mean',
        'Pases_recibidos': 'mean',
        'Pases_progresivos_recibidos': 'mean',
        'Faltas_cometidas': 'mean',
        'Duelos_aereos_ganados': 'mean'
    })
    .reset_index()
)

# Merge final
df_final = df_promedios.merge(df_ultimo_valor, on="Jugador", how="left")



df_final.head()
```
